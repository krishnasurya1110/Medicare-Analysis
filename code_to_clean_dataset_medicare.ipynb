```python
# How the original dataset was cleaned
import pandas as pd
medicare = pd.read_csv("medicare_2013_to_2023.csv")
medicare.columns = [col.lower() for col in medicare.columns]
medicare = medicare[medicare['bene_county_desc'] != 'Total']
medicare = medicare[medicare['month'] != 'Year']
medicare = medicare[medicare['bene_geo_lvl'] != 'State']
medicare = medicare[medicare['year'] != 2023] # since we dont have data for December 2023
medicare = medicare[medicare['bene_county_desc'] != 'Unknown']
medicare.head()

# Presence of *'s in almost all the columns
for i in medicare.columns:
    if '*' in medicare[i].value_counts():
        # print(i)
        pass

# Removing *'s
print(len(medicare))
medicare_cleaned = medicare[~medicare.isin(['*']).any(axis=1)]
medicare_cleaned.reset_index(drop=True, inplace=True)
print(len(medicare_cleaned))

# Percentage of data retained after dropping rows with *'set
print(round(len(medicare_cleaned)/ len(medicare), 2))

# Function to map state to region
def map_state_to_region(state):
    for region, states in mapping_dict.items():
        if state in states:
            return region
    return state

mapping_dict = {
    'Northeast': ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 'Rhode Island', 'Vermont',
                  'New Jersey', 'New York', 'Pennsylvania'],
    'Midwest': ['Illinois', 'Indiana', 'Michigan', 'Ohio', 'Wisconsin',
                'Iowa', 'Kansas', 'Minnesota', 'Missouri', 'Nebraska', 'North Dakota', 'South Dakota'],
    'South': ['Delaware', 'Florida', 'Georgia', 'Maryland', 'North Carolina', 'South Carolina', 'Virginia',
              'West Virginia', 'District of Columbia', 'Alabama', 'Kentucky', 'Mississippi', 'Tennessee', 'Arkansas', 'Louisiana',
              'Oklahoma', 'Texas'],
    'West': ['Arizona', 'Colorado', 'Idaho', 'Montana', 'Nevada', 'New Mexico', 'Utah', 'Wyoming',
             'Alaska', 'California', 'Hawaii', 'Oregon', 'Washington'],
    'Others': ['Puerto Rico', 'Virgin Islands']
}

medicare_cleaned = medicare_cleaned.copy()

# medicare_cleaned.loc[:, 'region'] = medicare_cleaned['bene_state_desc'].map(map_state_to_region)
region_column = medicare_cleaned['bene_state_desc'].map(map_state_to_region)
medicare_cleaned.insert(2, 'region', region_column)

medicare_cleaned.drop(columns=['bene_geo_lvl'], inplace=True) # Same entry for all rows 'County'

medicare_cleaned.rename(columns=lambda x: x.replace('bene_', '').replace('benes_', '').replace('_benes', '').replace('benes', ''), inplace=True)
medicare_cleaned.rename(columns=lambda x: x.replace('tot', 'total'), inplace=True)
medicare_cleaned.rename(columns=lambda x: x.replace('_mdcr', ''), inplace=True)

medicare_cleaned.rename(columns={
                                'state_abrvtn': 'state_code',
                                'state_desc': 'state',
                                'county_desc': 'county',
                                'fips_cd': 'fips_code',
                            }, inplace=True)

# # Save the cleaned dataset
medicare_cleaned.to_csv('medicare_cleaned.csv', index = False)

print(medicare_cleaned.columns)

pd.set_option('display.max_columns', None)
medicare_cleaned.head()
```
