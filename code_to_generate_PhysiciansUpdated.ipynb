```python
import pandas as pd
df = pd.read_csv('n_physicians_data_now.csv')
df.head()
```

```python
print(df['Rfrg_Prvdr_Cntry'].value_counts())

# Filter data for US
df = df[df['Rfrg_Prvdr_Cntry'] == 'US']
```

```python
print(df['Rfrg_Prvdr_Crdntls'].value_counts())

# Remove . inbetween letters to avoid duplicates
df['Rfrg_Prvdr_Crdntls'] = df['Rfrg_Prvdr_Crdntls'].str.replace('.', '')

print(df['Rfrg_Prvdr_Crdntls'].value_counts())
```

```python
# Replace a few more credentials
replace_dict = {', MD, PHD': 'MD, PHD',
                ', PA-C': 'PA-C',
                ',PA-C': 'PA-C',
                '(PA)': 'PA',}

df['Rfrg_Prvdr_Crdntls'] = df['Rfrg_Prvdr_Crdntls'].replace(replace_dict)
```

```python
# Rename columns
print(df.columns)

df.drop(columns=['Rfrg_Prvdr_MI', 'Rfrg_Prvdr_St1', 'Rfrg_Prvdr_St2', 'Rfrg_Prvdr_State_FIPS',
                 'Rfrg_Prvdr_RUCA', 'Rfrg_Prvdr_RUCA_Desc',
                 'Rfrg_Prvdr_Type_cd', 'Rfrg_Prvdr_Type_Flag',
                 'Rfrg_Prvdr_Cntry'], inplace=True)

df.rename(columns=lambda x: x.replace('Rfrg_', '').replace('Prvdr_', ''), inplace=True)
df.rename(columns=lambda x: x.replace('_cd', 'code').replace('_CD', '_code').replace('_Cd', '_code').replace('_Desc', '_desc'), inplace=True)

df.rename(columns={'First_Name': 'FirstName',
                    'Last_Name_Org': 'LastName',
                    'Crdntls': 'Credentials',
                    'Gndr': 'Gender',
                    'Ent': 'Entity',
                    'Zip5': 'Zipcode',
                    'Crdntls': 'Credentials',
                    'Gndr': 'Gender',
                    'Ent_code': 'Entity_code',
                    'RUCA_CAT': 'RUCA_category',
                    'BETOS_Lvl': 'BETOS_level',
                    }, inplace=True)

print(df.columns)
```
