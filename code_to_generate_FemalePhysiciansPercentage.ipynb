```python
import pandas as pd
df = pd.read_csv('n_physicians_only.csv')
df.head()
```

```python
result = df.groupby('State').agg(
    F_count=('Gender', lambda x: (x == 'F').sum()),  # Count of females
    ID_count=('NPI', 'count')  # Count of IDs
)
result['FemalePercentage'] = round((result['F_count']/ result['ID_count'])* 100, 2)
result.reset_index(inplace=True)
result.head()
```

```python
result.to_csv('n_female_percentage.csv', index=False)
```
